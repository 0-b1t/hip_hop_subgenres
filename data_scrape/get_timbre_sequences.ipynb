```python
def get_majority_timbre_vecs(t_json):
    seg_tatum_idxs = []
    tatum_starts = [x["start"] for x in t_json["tatums"]]
    for i in range(len(tatum_starts) - 1):
        tatum_begin = tatum_starts[i]
        tatum_end = tatum_starts[i + 1]
        segs_in_range = []
        for j in range(len(t_json["segments"])):
            seg_begin = t_json["segments"][j]["start"]
            seg_end = seg_begin + t_json["segments"][j]["duration"]
            if seg_begin > tatum_end or seg_end < tatum_begin:
                pass
            elif seg_begin < tatum_end or (seg_end > tatum_begin):
                segs_in_range.append(j)

        seg_tatum_idxs.append(segs_in_range)
    timbre_vecs = []
    for i in range(len(tatum_starts) - 1):
        #     print(tatum_starts[i])
        tatum_seg_ratios = []
        tatum_dur = tatum_starts[i + 1] - tatum_starts[i]
        for seg_idx in seg_tatum_idxs[i]:
            seg = t_json["segments"][seg_idx]
            seg_begin = seg["start"]
            seg_dur = seg["duration"]
            seg_end = seg_begin + seg_dur
            # subtract any of the duration that comes before the beginning of the tatum
            if seg_begin < tatum_starts[i]:
                seg_dur -= tatum_starts[i] - seg_begin
            # same for the end of the tatum
            if seg_end > tatum_starts[i + 1]:
                seg_dur -= seg_end - tatum_starts[i + 1]
            # store it as a ratio to see which segment is the majority of the tatum
            seg_ratio = seg_dur / tatum_dur
            #         print(t_json["segments"][seg_idx])
            tatum_seg_ratios.append(seg_ratio)
        majority_seg_idx = tatum_seg_ratios.index(max(tatum_seg_ratios))
        majority_seg = seg_tatum_idxs[i][majority_seg_idx]
        #     print(majority_seg)
        timbre_vec = t_json["segments"][majority_seg]["timbre"]
        timbre_vecs.append(timbre_vec)

    return timbre_vecs
```

```python
locked_sequences = {}
load_bar = 50
count = 0
for t in tracks.index:
    #     t = tracks.loc[i,'id']
    with open(f"../data/analysis_jsons/{t}.json", "r") as f:
        t_json = json.load(f)
    timbre_vecs = get_majority_timbre_vecs(t_json)
    locked_sequences[t] = timbre_vecs
    count += 1
    if count % load_bar == 0:
        print((count // load_bar) * load_bar)
```

```python
with open("../data/sequ_locked.pickle", "wb") as f:
    pickle.dump(locked_sequences, f)
```
