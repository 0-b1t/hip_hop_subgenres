```python
track_info = []

import time

st = 0
end = len(sp_ids)
# end = 200
step = 50

list(range(st, end, step))
for i in range(st, end, step):
    print(i)
    if len(sp_ids) - i >= step - 1:
        response = sp.tracks(sp_ids[i : i + step])
    else:
        response = sp.tracks(sp_ids[i:])
    time.sleep(1)
    track_info.append(response)
```

```python
info_dict = {}
for batch in track_info:
    type(batch)
    for ind in batch["tracks"]:
        track_artist = ind["artists"][0]["name"]

        artist_ids = ", ".join([x["id"] for x in ind["artists"]])

        track_name = ind["name"]

        track_album = ind["album"]["name"]

        track_popularity = ind["popularity"]

        track_id = ind["id"]

        info_dict[track_id] = [
            track_artist,
            track_name,
            track_album,
            track_popularity,
            artist_ids,
        ]
len(info_dict)
```

```python
tdf = pd.DataFrame(info_dict).T
tdf = tdf.reset_index()
# tdf.columns

tdf = tdf.rename(
    columns={
        "index": "id",
        0: "artist",
        1: "title",
        2: "album",
        3: "popularity",
        4: "artist_ids",
    }
)
tdf.head()
```

```python
track_features = []

import time

st = 0
end = len(sp_ids)
# end = 200
step = 100

list(range(st, end, step))
for i in range(st, end, step):
    print(i)
    if len(sp_ids) - i >= step - 1:
        response = sp.audio_features(sp_ids[i : i + step])
    else:
        response = sp.audio_features(sp_ids[i:])
    time.sleep(0.3)
    track_features.append(response)
```
