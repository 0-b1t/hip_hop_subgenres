```python
permutations = ["rap", "hip hop", "hip-hop", "hiphop"]
```

```python
# g = "alternative rap"
def return_permuts(g):
    if g.split(" ")[-1] == "rap":
        perms = [" ".join(g.split(" ")[:-1] + [p]) for p in permutations]
    else:
        perms = [" ".join(g.split(" ") + [p]) for p in permutations]
        perms.append(g)
    return perms
```

```python
permutation_dict = {}
for s in subgenres:
    permutation_dict[s] = return_permuts(s)
```

```python
perm_pre_df = []
for s in subgenres:
    for p in permutation_dict[s]:
        tag = network.get_tag(p)
        artists = tag.get_top_artists(limit=30)
        for a in artists:
            artist = a.item
            name = artist.name
            playcount = artist.get_playcount()
            item_dict = {"artist": name, "tag": s, "plays": playcount}
            perm_pre_df.append(item_dict)
```

```python
def save_song_lyrics(genius_artist):
    artist_name = artist.name
    artist_name = artist_name.replace('/', '_')
    path = f"../data/song_lyrics/{artist_name}"
    if not os.path.isdir(path):
        os.mkdir(path)
        for s in artist.songs:
            song_title = s.title
            song_title=song_title.replace('/', '_')
            with open(
                f"../data/song_lyrics/{artist_name}/{song_title}.txt", "w"
            ) as text_file:
                text_file.write(s.lyrics)
```

```python
def check_existence(artist_search):
    artist = genius.search_artist(
        artist_search, max_songs=1, sort="popularity", allow_name_change=False
    )
    if artist != None:
        artist_name = artist.name
        artist_name = artist_name.replace("/", "_")
        path = f"../data/song_lyrics/{artist_name}"
        return os.path.isdir(path)
    elif artist == None:
        return False
```
