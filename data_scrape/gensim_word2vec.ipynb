```python
def preprocess(doc):
    stemmer = SnowballStemmer("english")
    wnl = WordNetLemmatizer()
    doc = doc.replace("\n", " ")
    tokenized = word_tokenize(doc)

    cleaned = [
        #         wnl.lemmatize(stemmer.stem(token.lower()))
        #         wnl.lemmatize(token.lower())
        token.lower()
        for token in tokenized
        if not token.lower() in stopwords.words("english")
        if token.isalpha()
    ]
    untokenized = " ".join(cleaned)
    return untokenized
```

```python
filtered_lyrics["cleaned_3"] = filtered_lyrics["original"].apply(
    lambda x: preprocess(x)
)
```

```python
filtered_lyrics["cleaned_3"]
```

```python
filtered_lyrics.to_csv('../datacleaned_3_lyrics.csv')
```

```python
docs = filtered_lyrics["cleaned_3"].values
documents = [d.split(" ") for d in docs]
```

```python
%reload_ext nb_black
```

```python
my_model = gensim.models.Word2Vec(
    documents, size=200, window=5, min_count=1, workers=4, sg=0, iter=10
)
```

```python
my_model.most_similar("lightin")
my_model.most_similar("pregame")
```

```python
lyric_vectors = {}
for i in filtered_lyrics.index:
    l = filtered_lyrics.loc[i, "cleaned_3"]
    l_vec = [list(my_model[w]) for w in l.split(" ")]
    lyric_vectors[i] = l_vec
```
